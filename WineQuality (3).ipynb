# Dataset Preprocesssing

```
#from what range value does quality column has
print(data['quality'].unique())
```

```
from sklearn.preprocessing import LabelEncoder

# Assuming data is your DataFrame
# Map quality values 3, 4, 5 to 'bad' and others to 'good'
data['quality'] = data['quality'].apply(lambda x: 'bad' if x in [3, 4, 5] else 'good')

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Apply LabelEncoder to the quality column
data['quality'] = label_encoder.fit_transform(data['quality'])

print(data['quality'].unique())
```

Mapping Quality Ratings: The quality column is reclassified as 'bad' for ratings 3, 4, and 5, and 'good' for ratings 6, 7, and 8. This simplifies the target variable into two categories.

Label Encoding: The LabelEncoder converts 'bad' to 0 and 'good' to 1, transforming the quality column into binary values.

The unique values [0, 1] in the output confirm that quality is now encoded as a binary classification variable.

Train and test the model with cross validation

```
# dividing the dataset into dependent and independent variables

x = data.iloc[:,:11]
y = data.iloc[:,11]

# determining the shape of x and y.
print(x.shape)
print(y.shape)
```

```
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 40)

x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

Splitting the Data: The dataset is split into training and testing sets, with 25% of the data allocated for testing and 75% for training. The random_state=40 ensures reproducibility of the split.

Shapes: The output shapes indicate:

x_train: 1199 samples with 11 features (independent variables).
y_train: 1199 samples for the target variable.
x_test: 400 samples with 11 features.
y_test: 400 samples for the target variable.
This split will allow you to train the model on 75% of the data and evaluate its performance on the remaining 25%.

# **Modelling**

**Logistic Regression**

```
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

from sklearn.model_selection import GridSearchCV, cross_val_score



# creating the model
model = LogisticRegression(solver='lbfgs', max_iter=1000)


# feeding the training set into the model
model.fit(x_train, y_train)

# predicting the results for the test set
y_pred = model.predict(x_test)

# calculating the training and testing accuracies
print("Training accuracy :", model.score(x_train, y_train))
print("Testing accuracy :", model.score(x_test, y_test))

# classification report
print(classification_report(y_test, y_pred))

# confusion matrix
print(confusion_matrix(y_test, y_pred))
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Train the LogisticRegression model
# Generate probability estimates for the positive class
y_proba = model.predict_proba(x_test)[:, 1]  # Get probabilities for class 1

# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_proba)

# Calculate AUC
roc_auc = roc_auc_score(y_test, y_proba)

# Plot ROC curve
plt.figure()
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # Diagonal line for random performance
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

```
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.metrics import accuracy_score

# Hyperparameters for each model
models = {
    "Decision Tree": DecisionTreeClassifier(max_depth=5, min_samples_split=10),
    "K-Nearest Neighbors": KNeighborsClassifier(n_neighbors=5, weights='distance'),
    "Random Forest": RandomForestClassifier(n_estimators=100, max_depth=10, min_samples_split=10),
    "Logistic Regression": LogisticRegression(C=1.0, solver='liblinear', max_iter=1000),
    "SGD Classifier": SGDClassifier(learning_rate='optimal', penalty='l2', max_iter=1000)
}

# Initialize lists to store results for the table
model_names = []
train_accuracies = []
test_accuracies = []

# Train each model and compute accuracies
for name, model in models.items():
    model.fit(x_train, y_train)  # Train the model

    # Calculate training and testing accuracy
    train_accuracy = accuracy_score(y_train, model.predict(x_train))
    test_accuracy = accuracy_score(y_test, model.predict(x_test))

    # Append results to lists
    model_names.append(name)
    train_accuracies.append(train_accuracy)
    test_accuracies.append(test_accuracy)

# Create a DataFrame for the results
results_df = pd.DataFrame({
    "Model": model_names,
    "Training Accuracy": train_accuracies,
    "Testing Accuracy": test_accuracies
})

# Display the table
print("Training and Testing Accuracy Comparison:")
print(results_df)
```

Random Forest appears to be the best model here, with the highest test accuracy (0.8300), despite slight overfitting.
Logistic Regression might be a good choice if you want a simpler model that generalizes without much overfitting. It has a balanced accuracy and could be preferred for interpretability and lower complexity.
In summary, Random Forest is the top performer in terms of accuracy, while Logistic Regression may provide a good balance between simplicity and performance.

### **Apply Hyperparameter Tunning  for Random forest and Logistic regression**


```
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score
import numpy as np

# Define a parameter distribution for RandomForest
rf_param_dist = {
    'n_estimators': [50, 100, 150],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'bootstrap': [True, False]
}

# Initialize the RandomForestClassifier
rf_model = RandomForestClassifier()

# Set up RandomizedSearchCV with 10 iterations
rf_random_search = RandomizedSearchCV(estimator=rf_model, param_distributions=rf_param_dist,
                                      n_iter=10, cv=3, scoring='accuracy', n_jobs=-1, random_state=42, verbose=1)

# Fit RandomizedSearchCV to the training data
rf_random_search.fit(x_train, y_train)

# Print the best parameters and best cross-validation score
print("Best parameters for Random Forest:", rf_random_search.best_params_)
print("Best cross-validation accuracy for Random Forest:", rf_random_search.best_score_)

# Get the testing accuracy for the best model
best_rf_model = rf_random_search.best_estimator_
rf_test_accuracy = accuracy_score(y_test, best_rf_model.predict(x_test))
print(f"Testing Accuracy for Random Forest with best parameters: {rf_test_accuracy:.4f}")
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV
import numpy as np

# Define a smaller parameter distribution for LogisticRegression without None penalty
lr_param_dist = {
    'penalty': ['l2'],  # Only use 'l2' for this solver
    'C': np.logspace(-2, 2, 5),  # Include a range of values for C
    'solver': ['liblinear'],  # Fast solver suitable for small datasets
    'max_iter': [100, 200]
}

# Initialize the LogisticRegression model
lr_model = LogisticRegression()

# Set up RandomizedSearchCV with 10 iterations
lr_random_search = RandomizedSearchCV(estimator=lr_model, param_distributions=lr_param_dist,
                                      n_iter=10, cv=3, scoring='accuracy', n_jobs=-1, random_state=42, verbose=1)

# Fit RandomizedSearchCV to the training data
lr_random_search.fit(x_train, y_train)

# Print the best parameters and best score
print("Best parameters for Logistic Regression:", lr_random_search.best_params_)
print("Best cross-validation accuracy for Logistic Regression:", lr_random_search.best_score_)

# Get the testing accuracy for the best model
best_lr_model = lr_random_search.best_estimator_
lr_test_accuracy = accuracy_score(y_test, best_lr_model.predict(x_test))
print(f"Testing Accuracy for Logistic Regression with best parameters: {lr_test_accuracy:.4f}")
```
